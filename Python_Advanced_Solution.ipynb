```python
#only using the data we need
used_cols = ['title', 'date_added', 'release_year', 'rating', 'duration', 'description', 'listed_in', 'cast']
data = data[used_cols]
data
```

# preparing the data for the Rec System

```python
# first it is important to delete capital letters and 
# replace them with lowercase letters (needed to create word-vectors)
# since it is a technique from natural language processing, we use the 
# nltk library (https://www.nltk.org ; possible to install in Anaconda)

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation
from nltk.stem.snowball import SnowballStemmer

data['title'] = data['title']
data['description'] = data['description'].str.lower()
data['listed_in'] = data['listed_in'].str.lower()
#data['cast'] = data['cast'].str.lower()
#data['cast'] = data['cast'].apply(lambda x: y.replace(' ', '') for y in x]])

# using a stemmer to change words with similar meaning to the same value
stemmer = SnowballStemmer('english')

# in NLP filler words like and, the, etc. are removed because the lack
# overall meaning and value for the text
# adding punctuation to the filler words
rem_words = stopwords.words('english') + list(punctuation)

# tokenizing the different sentences to a bag of words
# changing words with a comparable meaning to the same word (->stemming)
# words to remove the stop words
# needs to be done(partially) for description, actor, listed_in

# description
# tonenize the single words in the description / create vectors out of them
data['description'] = data['description'].apply(word_tokenize)
# apply stemming
data['description'] = data['description'].apply(lambda x: [stemmer.stem(y) for y in x])
# remove the filler words that do not add meaning to the sentences
data['description'] = data['description'].apply(lambda x: ' '.join([word for word in x if word not in (rem_words)]))

# cast
# replacing the NaN
data['cast'] = data['cast'].fillna('unknown') # necessary since not every film has a cast listed in the dataset
# concatenating the names of the cast member and replacing the comma with a space
data['cast'] = data['cast'].apply(lambda x: x.replace(' ', ''))
data['cast'] = data['cast'].apply(lambda x: x.replace(',', ' '))
# tokenizing the string of cast members
#data['cast'] = data['cast'].apply(word_tokenize)

# genres (listed_in)
# concatenating the names of the cast member and replacing the comma with a space
data['listed_in'] = data['listed_in'].apply(word_tokenize)
data['listed_in'] = data['listed_in'].apply(lambda x: " ".join(x))
data['listed_in'] = data['listed_in'].apply(lambda x: x.replace(' ', ''))
data['listed_in'] = data['listed_in'].apply(lambda x: x.replace(',', ' '))
# tokenizing the string of genres
#data['listed_in'] = data['listed_in'].apply(word_tokenize)
#data['listed_in'] = data['listed_in'].apply(lambda x: ' '.join([word for word in x if word not in (rem_words)]))
```

```python
# constructing the Term-Document Matrix

# Activate CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

# Count Vectorizer
# all the vectors for every description and how often they appear are put in a variable
vect = CountVectorizer()  
vects = vect.fit_transform(data['description'])
```

```python
# term document matrix for description
tdm_descr = pd.DataFrame(vects.todense())
tdm_descr.columns = vect.get_feature_names()
#tdm_descr = tdm_descr.T not necessary when using the python modules
tdm_descr.index = [title for title in data['title']]
#tdm_descr['total_count'] = tdm_descr.sum(axis=1)
tdm_descr
```

```python
# importing the cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# using the cosine similarity on the tdm of the relevant features
cos_sim_descr = cosine_similarity(tdm_descr)
# putting the resulting matrix into a dataframe -> easier to use it further
cs_descr = pd.DataFrame(cos_sim_descr)
# naming the columns and index after the index of the tdm <- didn't change by using the function
cs_descr.columns = cs_descr.index = tdm_descr.index
cs_descr
```

```python
# now the same for genre
vect = CountVectorizer()  
vects = vect.fit_transform(data['listed_in'])
tdm_genre = pd.DataFrame(vects.todense())
tdm_genre.columns = vect.get_feature_names()
#tdm_genre = tdm_genre.T not necessary when using the python modules
tdm_genre.index = [title for title in data['title']]
#tdm_genre['total_count'] = tdm_genre.sum(axis=1)
tdm_genre
```

```python
#same steps again
cos_sim_genre = cosine_similarity(tdm_genre)
cs_genre = pd.DataFrame(cos_sim_genre)
cs_genre.columns = cs_genre.index = tdm_genre.index
cs_genre
```

```python
# now the same for actors
vect = CountVectorizer()  
vects = vect.fit_transform(data['cast'])
tdm_cast = pd.DataFrame(vects.todense())
tdm_cast.columns = vect.get_feature_names()
#tdm_cast = tdm_cast.T not necessary when using the python modules
tdm_cast.index = [title for title in data['title']]
#tdm_genre['total_count'] = tdm_genre.sum(axis=1)
tdm_cast
```

```python
#same steps again
cos_sim_cast = cosine_similarity(tdm_cast)
cs_cast = pd.DataFrame(cos_sim_cast)
cs_cast.columns = cs_cast.index = tdm_cast.index
cs_cast
```

# now creating the recommendation function

```python
w_genre = 0.2
w_descr = 0.3
w_cast = 0.4
# must add up to one

def recommendation(movie: str, w_genre, w_descr, w_cast):
    if movie in cs_genre.columns:
        test_genre = pd.DataFrame()
        test_genre['genre']=cs_genre[movie]

        test_actor = pd.DataFrame()
        test_actor['cast']=cs_cast[movie]

        test_descr = pd.DataFrame()
        test_descr['description']=cs_descr[movie]


        df_final = pd.concat([test_descr, test_actor, test_genre], ignore_index=False)
        df_final = df_final.fillna(0) #after concat, there are NaNs in the dataframe again
        df_final['result'] = (w_descr/(w_genre+w_descr+w_cast))*df_final['description'] + (w_genre/(w_genre+w_descr+w_cast))*df_final['genre'] + (w_cast/(w_genre+w_descr+w_cast))*df_final['cast']
        return df_final
    
    else:
        pass
        return None
```

```python
for i in tobi_data['title_new']:
    recs = recommendation(i, w_genre, w_descr, w_cast)
    if recs is not None:
        print('for ', i,' your recommendations are: ')
        recs = recs.sort_values('result', ascending=False)
        print(recs['result'][2:11])
    else:
        pass
```
